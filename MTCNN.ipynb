```
def draw_box_in_image(file, res):
    data = pyplot.imread(file)
    pyplot.imshow(data)
    ax = pyplot.gca()
    for i in res:
        x, y, w, h = i['box']
        rect = Rectangle((x,y), w, h, fill = False, color = 'red')
        ax.add_patch(rect)
    pyplot.show()
```

```
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from mtcnn.mtcnn import MTCNN
from PIL import Image
from numpy import asarray
```

```
file = 'fm.jpg'

detector = MTCNN()

pix = pyplot.imread(file)
```

```
faces = detector.detect_faces(pix)

draw_box_in_image(file, faces)
```

**Extract faces from images**

```
def extract_face(file):
    pic = pyplot.imread(file)
    detect = MTCNN()
    res = detect.detect_faces(pic)
    for i in range(len(res)):
        
        x1, y1, w, h = res[i]['box']
        x2, y2 = x1+w, y1+h
        
        pyplot.subplot(1, len(res), i+1)
        pyplot.axis('off')
        
        pyplot.imshow(pic[y1:y2, x1:x2])
    
    pyplot.show()
```

```
file = 'uzl.jpg'

detector = MTCNN()

pix = pyplot.imread(file)

faces = detector.detect_faces(pix)

draw_box_in_image(file, faces)
```

```
extract_face(file)
```
